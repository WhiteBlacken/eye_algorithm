```python
import pandas as pd
data = pd.read_csv("data/raw_data_0317.csv")
data
```

```python
from nltk.corpus import brown
brown_words = brown.words()
def word_frequency(brown_words, word):
    # 计算目标词在Brown语料库中的出现频率
    return brown_words.count(word)
```

```python
word_to_freq = {}
```

```python
for i, row in data.iterrows():
    text_sequences = eval(row['text_sequences'])
    for word_list in text_sequences:
        word = word_list[1]
        word_list.append(len(word))
        if word in word_to_freq:
            freq = word_frequency[word]
        else:
            freq = word_frequency(brown_words, word)
        word_list.append(freq)
    print(text_sequences)
    break
```

```python
page_skip_info = {}
for i, row in data.iterrows():
    page_skip_info[row['page_id']] = len(row['eye_tracking_sequences'])/len(row['text_sequences'])

fixations_per_words = [val for key, val in page_skip_info.items()]
fixations_per_words.sort()

min_no_skip_thredhold = fixations_per_words[int(len(fixations_per_words)*0.4)]
max_skip_thredhold = fixations_per_words[int(len(fixations_per_words)*0.6)]

min_no_skip_thredhold, max_skip_thredhold

skip_pages = [key for key, val in page_skip_info.items() if val < max_skip_thredhold]
not_skip_pages = [key for key, val in page_skip_info.items() if val > min_no_skip_thredhold]
len(skip_pages), len(not_skip_pages)
```

```python
data[data['page_id'].isin(skip_pages)].to_csv('data/skip_data.csv')
data[data['page_id'].isin(not_skip_pages)].to_csv('data/not_skip_data.csv')
```
