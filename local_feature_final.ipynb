```python
import pandas as pd

# 读取 csv 文件
data = pd.read_csv('your_file.csv')

# 定义要处理的眼动指标列
eye_movement_columns = ['number_of_fixations', 'fixation_duration', 'reading_times']

# 计算左右各两个值的平均值和最大值，并添加为新列
for column in eye_movement_columns:
    data[f'{column}_avg_left_two'] = data[column].rolling(window=2).mean().shift(-1)
    data[f'{column}_max_left_two'] = data[column].rolling(window=2).max().shift(-1)
    data[f'{column}_avg_right_two'] = data[column].rolling(window=2).mean().shift(1)
    data[f'{column}_max_right_two'] = data[column].rolling(window=2).max().shift(1)

# 将处理后的数据写入新的 csv 文件
data.to_csv('new_file.csv', index=False)
```
