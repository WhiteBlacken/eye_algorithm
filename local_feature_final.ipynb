```python
import pandas as pd
not_corrected_data_skip = pd.read_csv("data/final_data/word-feature-without-correct-skip_1.csv") # 无纠正跳读
not_corrected_data_skip
```

```python
import pandas as pd

# 读取 csv 文件
data = not_corrected_data_skip

# 定义要处理的眼动指标列
eye_movement_columns = ['number_of_fixations', 'fixation_duration', 'reading_times']

# 计算左右各两个值的平均值和最大值，并添加为新列
for column in eye_movement_columns:
    data[f'{column}_avg_left_two'] = data[column].rolling(window=2).mean().shift(-1)
    data[f'{column}_max_left_two'] = data[column].rolling(window=2).max().shift(-1)
    data[f'{column}_avg_right_two'] = data[column].rolling(window=2).mean().shift(1)
    data[f'{column}_max_right_two'] = data[column].rolling(window=2).max().shift(1)
    
    # # 合并左右的平均值
    # data[f'{column}_avg_combined'] = (data[f'{column}_avg_left_two'] + data[f'{column}_avg_right_two']) / 2
    # # 合并左右的最大值
    # data[f'{column}_max_combined'] = pd.concat([data[f'{column}_max_left_two'], data[f'{column}_max_right_two']]).max()
data.fillna(0, inplace=True)
# 将处理后的数据写入新的 csv 文件
data.to_csv('data/final_data/skip_window_data.csv', index=False)
```
