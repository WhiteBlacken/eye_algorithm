```python
import pandas as pd
data = pd.read_csv('data/raw_data_0317.csv')
data
```

```python
# 先拿一个当例子
fixation_sequence_for_one = []
text_sequence_for_one = []
now_y_loc = 0
for idx, row in data.iterrows():
    text_sequence = eval(row['text_sequences'])
    tmp = []
    for i, sequenece in enumerate(text_sequence):
        if i == 0:
            now_y_loc = sequenece[3]
        if sequenece[3] != now_y_loc:
            text_sequence_for_one.append([x for x in tmp])
            tmp = []
        tmp.append([x for x in sequenece])
        # print(tmp)
        now_y_loc = sequenece[3]
    break
print(text_sequence_for_one)
```

```python
fixation_sequence = eval(data.iloc[0]['eye_tracking_sequences'])
fixation_sequence
```

```python
# 注视也拆分序列
for sequenece in fixation_sequence:
    print(sequenece)
```

```python
import math
# 注视也拆分序列
velocity_list = []
for i, sequenece in enumerate(fixation_sequence): # x,y,t,vel
    if i == 0: continue
    now_fix = fixation_sequence[i]
    pre_fix = fixation_sequence[i-1]
    distance = math.sqrt(math.pow(now_fix[0]-pre_fix[0], 2) + math.pow(now_fix[1]-pre_fix[1], 2))
    velocity_list.append(distance/(now_fix[2]-pre_fix[2]))
print(velocity_list)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(velocity_list, kde=True)  # kde=True 表示同时绘制核密度估计图
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram and KDE Plot')
plt.show()
```
